### **Import Needed Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

### **Read Data and Apply Preprocessing**

```python
df = pd.read_csv('Position_Salaries.csv')
df
```

**Define Features X and Target y**

```python
X = df.iloc[:, 1:2].values
y = df.iloc[:, 2].values
```

**Visualize the data**

```python
plt.figure(figsize=(10, 5))

plt.scatter(X, y, color='blue')
plt.title('Positions Salaries')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.grid(True)

plt.show()
```

**Convert Data into Polynomial with degree**

```python
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
```

**Apply Linear Regression Model**

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

**Visualising the Linear Regression results**

```python
plt.figure(figsize=(10, 5))

plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg.predict(X), color='red')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.grid(True)

plt.show()
```

**Apply Linear Regression with Ploynomial Data**

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

**Visualising the Polynomial Regression results**

```python
plt.figure(figsize=(10, 5))

plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'red')
plt.title('(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.grid(True)

plt.show()
```

**Predicting a new result with Linear Regression**

```python
lin_reg.predict([[6.5]])
```

**Predicting a new result with Polynomial Regression**

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
